# Calculo para el Aprendizaje Automático

Hasta ahora venimos trabajando con el algorito K-Nearest Neighbors, aunque el algoritmo es facíl de entender, no es posible utilizarlo para conjuntos de datos muy grandes, esto debido a su propio funcionamiento. Ya que cada vez que queremos hacer una predicción, tenemos que calcular la distancia entre cada observacion de el conjunto de entrenamiento y la nueva observación, y luego ordenar por distancia ascendente. Esto es muy intensivo desde el punto de vista informático.<br>

En la mayoria de las técnicas de ML que se veran a continuación, el modelo _**se representa como una función matemática**_. Esta, se aproxima a la función subyacente que describe cómo se relacionan las caracteristicas con el atributo objetivo. De esta manera, hacer predicciones es computacionalmente barato.<br>

_El cáculonos proporicona un marco para entender cómo se comportan las funciones matemáticas_:
- Entender la inclinación en varios puntos.
- Encontrar los puntos extremos de una funcion.
- Determinar la función óptima que mejor representa un conjunto de datos.

Supongamos que se nos da la siguiente ecuación, la cual describe la trayectoria de un balon despues de ser pateado: $$ y = -x^{2} + 3x - 1 $$

Donde:
- $x$ es el tiempo, en segundos.
- $y$ es la posición vertical de la pelota.

Naturalmente, nos gustaria saber la posicion mas alta que alcanzo la pelota. Aunque podemos graficar y estimar el resultado visualmente, si queremos el tiempo y la posicion precisos tendremos que usar el cáculo.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 3, 100)
y = -1*(x**2) + (x*3) - 1

plt.plot(x,y)
plt.show()
```
