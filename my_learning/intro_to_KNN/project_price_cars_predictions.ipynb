# Proyecto guiado: Predicción de los Precios de Automóviles.

```python
import pandas as pd
import numpy as np

# Estableciendo un numero mayor para desplegar columnas
pd.options.display.max_columns = 99
```

```python
columns_name = ['symboling', 'normalized-losses', 'make', 'fuel_type', 'aspiration', 'num_doors', 'body-style', 'drive-wheels', 'engine-location', 'wheel-base', 
                'length', 'width', 'height', 'curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate',
                'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']

cars = pd.read_csv('./imports-85.data', names=columns_name)
cars.head(10)
```

```python
# Seleccionando columnas con valores numericos.
continuous_values_cols = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-size', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
numeric_cars = cars[continuous_values_cols]
numeric_cars.head(5)
```

### Data Cleaning

```python
numeric_cars = numeric_cars.replace('?', np.nan)
numeric_cars.head(5)
```

```python
numeric_cars = numeric_cars.astype('float')
numeric_cars.isnull().sum()
```

```python
# Como 'price' es la caracteristica que deseamos predecir, removeremos los que tengan nulos.
numeric_cars = numeric_cars.dropna(subset=['price'])
numeric_cars.isnull().sum()
```

```python
# Reemplazando los demas valores nulos por el promedio de la esa caracteristica (columna)
numeric_cars = numeric_cars.fillna(numeric_cars.mean())
numeric_cars.isnull().sum()
```

##### Normalizacion Min-Max [(vease)](https://jorgeiblanco.medium.com/por-qu%C3%A9-la-normalizaci%C3%B3n-es-clave-e-importante-en-machine-learning-y-ciencia-de-datos-4595f15d5be0#:~:text=La%20normalizaci%C3%B3n%20Min%2DMax%3A%20es,t%C3%ADpicamente%20entre%200%20y%201.)

Definida como: $$ N_{i} = \frac{ (X_{i} - X_{min}) }{ X_{max} - X_{min} } $$

```python
# Normalizando todas las columnas en un rango de 0 a 1.
price_col = numeric_cars['price']
numeric_cars = (numeric_cars - numeric_cars.min())/(numeric_cars.max() - numeric_cars.min())
numeric_cars['price'] = price_col
numeric_cars.head()
```

### Modelo _Univariante_.

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error


def knn_train(train_col, target_col, df):
    knn = KNeighborsRegressor()
    np.random.seed(1)

    # Ordenando aleatoriamente el DataFrame
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)

    # Obteniendo el indice de la mitad del DataFrame
    last_train_row = int(len(rand_df) / 2)

    # Separando el DataFrame para el conjunto de entrenamiento y el de prueba
    train_df = rand_df.iloc[0:last_train_row]
    test_df = rand_df.iloc[last_train_row:]

    # Ajustando modelo
    knn.fit(train_df[[train_col]], train_df[target_col])

    # Haciendo predicciones
    predictions = knn.predict(test_df[[train_col]])

    # Calculando RMSE y retornandolo
    mse = mean_squared_error(test_df[target_col], predictions)
    rmse = np.sqrt(mse)
    return rmse

rmse_results = {}
trains_cols = numeric_cars.columns.drop('price')

# Evaluando cada caracteristica (columna) a excepcion del 'price'.
for col in trains_cols:
    rmse = knn_train(col, 'price', numeric_cars)
    rmse_results[col] = rmse

# Convirtiendo dict de Python a Series de Pandas
rmse_results_series = pd.Series(rmse_results)
rmse_results_series.sort_values()
```

##### Optimizacion del hiperparametro _K_.

```python
def knn_train(train_col, target_col, df):
    np.random.seed(1)
    
    # Ordenando aleatoriamente los datos.
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)

    # Obteniendo el indice de en medio del DataFrame.
    last_train_row = int(len(rand_df) / 2)

    # Separando el DataFrame en conjunto de entrenamiento y prueba
    train_df = rand_df.iloc[0:last_train_row]
    test_df = rand_df.iloc[last_train_row:]

    # Optimizacion de parametros aplicando BUSQUEDA EN CUADRICULA
    k_values = [k+1 for k in range(5)]
    k_rmses = {}

    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors=k)

        # Ajustando modelo
        knn.fit( train_df[[train_col]], train_df[target_col] )

        # Haciendo predicciones
        predictions = knn.predict(test_df[[train_col]])

        # Calculando RMSE
        mse = mean_squared_error(test_df[target_col], predictions)
        rmse = np.sqrt(mse)

        k_rmses[k] = rmse
    
    return k_rmses

k_rmse_results = {}

# Evaluando cada caracteristica (columna) a excepcion del 'price'.
train_cols = numeric_cars.columns.drop('price')
for col in train_cols:
    rmse_values = knn_train(col, 'price', numeric_cars)
    k_rmse_results[col] = rmse_values

k_rmse_results
```

```python
###################################
# Visualizando resultados
###################################

import matplotlib.pyplot as plt

fig, axs = plt.subplots(layout='constrained', figsize=(8,5))

for k,v in k_rmse_results.items():
    x = list(v.keys())
    y = list(v.values())
    
    axs.plot(x,y, label=k)
    axs.set_xlabel('k value')
    axs.set_ylabel('RMSE')

fig.legend(loc='outside right', ncols=1, fontsize='large')
plt.show()
```

### Modelo _Multivariante_.

```python
############################################
# Calculando el promedio de RMSE y ordenando
# las mejorescategorias 
############################################

feature_avg_rmse = {}

for k,v in k_rmse_results.items():
    avg_rmse = np.mean( list(v.values()) )
    feature_avg_rmse[k] = avg_rmse

series_avg_rmse = pd.Series(feature_avg_rmse)
sorted_series_avg_rmse = series_avg_rmse.sort_values()
print(sorted_series_avg_rmse)
sorted_features = sorted_series_avg_rmse.index
sorted_features
```

```python
def knn_train_test(train_cols, target_col, df):
    np.random.seed(1)

    # Ordenando aleatoriamente los datos
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)

    # Seleccionando index del DataFrame para separarlo por la mitad
    last_train_row = int(len(rand_df) / 2)

    # Separando en conjunto de PRUEBA y conjunto de ENTRENAMIENTO.
    train_df = rand_df.iloc[0:last_train_row]
    test_df = rand_df.iloc[last_train_row:]

    # Seleccionando hiperparametros
    k_values = [5]
    k_rmse = {}

    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors=k)

        # Ajustando modelo
        knn.fit(train_df[train_cols], train_df[target_col])

        # Haciendo predicciones
        predictions = knn.predict(test_df[train_cols])

        # Evaluando modelo
        mse = mean_squared_error(test_df[target_col], predictions)
        rmse = np.sqrt(mse)
        k_rmse[k] = rmse

    return k_rmse

k_rmse_values = {}

for num_best_feats in range(2,7):
    k_rmse_values[ f'{num_best_feats} best features' ] = knn_train_test(
        sorted_features[:num_best_feats],
        'price',
        numeric_cars,
    )

k_rmse_values
```

##### Optimizacion del hiperparametro _K_.

```python
def knn_train_test(train_cols, target_col, df):
    np.random.seed(1)

    # Ordenando aleatoriamente los datos
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)

    # Obteniendo el indice de la mitad para separar en conjunto de entrenamiento y prueba
    last_train_row = int(len(rand_df) / 2)

    # Separando en conjunto de ENTRENAMIENTO y PRUEBA
    train_df = rand_df.iloc[0:last_train_row]
    test_df = rand_df.iloc[last_train_row:]

    # Estableciendo hiperparametros
    k_values = [ k for k in range(1,25) ]
    k_rmse = {}

    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors=k)

        # Ajustando modelo
        knn.fit(train_df[train_cols], train_df[target_col])
        
        # Haciendo predicciones
        predictions = knn.predict(test_df[train_cols])

        # Evaluando modelo
        mse = mean_squared_error(test_df[target_col], predictions)
        rmse = np.sqrt(mse)
        k_rmse[k] = rmse

    return k_rmse

k_rmse_results = {}

for num_best_feat in range(2,6):
    k_rmse_results[f'{num_best_feat} best features'] = knn_train_test(
        sorted_features[:num_best_feat],
        'price',
        numeric_cars,
    )

k_rmse_results
```

```python
#############################
# Visualizacion de los datos
#############################

fig, axs = plt.subplots(layout='constrained', figsize=(12,6))
plt.xticks([ n for n in range(1,25)] )

for k,v in k_rmse_results.items():
    x = list(v.keys())
    y = list(v.values())

    axs.plot(x, y, label=k)
    axs.set_xlabel('K value')
    axs.set_ylabel('RMSE')

fig.legend(loc='outside right', ncols=1, fontsize='large')
plt.show()
    
```
