# Proyecto guiado: Predicción de los Precios de Automóviles.

```python
import pandas as pd
import numpy as np

# Estableciendo un numero mayor para desplegar columnas
pd.options.display.max_columns = 99
```

```python
columns_name = ['symboling', 'normalized-losses', 'make', 'fuel_type', 'aspiration', 'num_doors', 'body-style', 'drive-wheels', 'engine-location', 'wheel-base', 
                'length', 'width', 'height', 'curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate',
                'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']

cars = pd.read_csv('./imports-85.data', names=columns_name)
cars.head(10)
```

```python
# Seleccionando columnas con valores numericos.
continuous_values_cols = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-size', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
numeric_cars = cars[continuous_values_cols]
numeric_cars.head(5)
```

### Data Cleaning

```python
numeric_cars = numeric_cars.replace('?', np.nan)
numeric_cars.head(5)
```

```python
numeric_cars = numeric_cars.astype('float')
numeric_cars.isnull().sum()
```

```python
# Como 'price' es la caracteristica que deseamos predecir, removeremos los que tengan nulos.
numeric_cars = numeric_cars.dropna(subset=['price'])
numeric_cars.isnull().sum()
```

```python
# Reemplazando los demas valores nulos por el promedio de la esa caracteristica (columna)
numeric_cars = numeric_cars.fillna(numeric_cars.mean())
numeric_cars.isnull().sum()
```

##### Normalizacion Min-Max [(vease)](https://jorgeiblanco.medium.com/por-qu%C3%A9-la-normalizaci%C3%B3n-es-clave-e-importante-en-machine-learning-y-ciencia-de-datos-4595f15d5be0#:~:text=La%20normalizaci%C3%B3n%20Min%2DMax%3A%20es,t%C3%ADpicamente%20entre%200%20y%201.)

Definida como: $$ N_{i} = \frac{ (X_{i} - X_{min}) }{ X_{max} - X_{min} } $$

```python
# Normalizando todas las columnas en un rango de 0 a 1.
price_col = numeric_cars['price']
numeric_cars = (numeric_cars - numeric_cars.min())/(numeric_cars.max() - numeric_cars.min())
numeric_cars['price'] = price_col
numeric_cars.head()
```

### Modelo _Univariante_.

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error


def knn_train(train_col, target_col, df):
    knn = KNeighborsRegressor()
    np.random.seed(1)

    # Ordenando aleatoriamente el DataFrame
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)

    # Obteniendo el indice de la mitad del DataFrame
    last_train_row = int(len(rand_df) / 2)

    # Separando el DataFrame para el conjunto de entrenamiento y el de prueba
    train_df = rand_df.iloc[0:last_train_row]
    test_df = rand_df.iloc[last_train_row:]

    # Ajustando modelo
    knn.fit(train_df[[train_col]], train_df[target_col])

    # Haciendo predicciones
    predictions = knn.predict(test_df[[train_col]])

    # Calculando RMSE y retornandolo
    mse = mean_squared_error(test_df[target_col], predictions)
    rmse = np.sqrt(mse)
    return rmse

rmse_results = {}
trains_cols = numeric_cars.columns.drop('price')

# Evaluando cada caracteristica (columna) a excepcion del 'price'.
for col in trains_cols:
    rmse = knn_train(col, 'price', numeric_cars)
    rmse_results[col] = rmse

# Convirtiendo dict de Python a Series de Pandas
rmse_results_series = pd.Series(rmse_results)
rmse_results_series.sort_values()
```
